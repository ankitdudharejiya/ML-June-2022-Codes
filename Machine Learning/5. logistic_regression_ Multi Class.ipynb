```python
import pandas as pd
```

```python
dataset = pd.read_csv('Iris.csv')
dataset.head()
```

```python
dataset
```

```python
dataset['Species'].unique()
```

```python
dataset['Species'].replace({'Iris-setosa' : '1', 'Iris-versicolor' :'2', 'Iris-virginica' : '3'} , inplace=True)
```

```python
dataset
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train, y_test =train_test_split(dataset[['SepalLengthCm' ,'SepalWidthCm' ,'PetalLengthCm', 'PetalWidthCm']], dataset['Species'] , 
                 test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train , y_train)
```

```python
X_test
```

```python
lr.predict(X_test)
```
